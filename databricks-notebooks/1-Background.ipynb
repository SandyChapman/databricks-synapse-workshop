# Background: Databricks + Synapse

## Azure Databricks

Azure Databricks is a data analytics platform optimized for the Microsoft Azure cloud services platform. Azure Databricks offers three environments for developing data intensive applications: **Databricks SQL**, **Databricks Data Science & Engineering**, and **Databricks Machine Learning**.

Databricks **Data Science & Engineering** (sometimes called simply "Workspace") is an analytics platform based on Apache Spark. It is integrated with Azure to provide one-click setup, streamlined workflows, and an interactive workspace that enables collaboration between data engineers, data scientists, and machine learning engineers. This workshop will exclusively forcus on Databricks Workspaces.


## Azure Synapse

Azure Synapse is an enterprise analytics service that accelerates time to insight across data warehouses and big data systems. Azure Synapse brings together the best of SQL technologies used in enterprise data warehousing, Spark technologies used for big data, Pipelines for data integration and ETL/ELT, and deep integration with other Azure services such as Power BI, CosmosDB, and AzureML.

* Data Warehouse (Dedicated SQL Pools)
* Data Lake Querying (Serverless SQL Pools)
* Spark (Spark Pools)

## So, what's the difference?

Similar in purpose, different in execution.

Synapse has a strong focus the powerful SQL Data Warehousing functionality and on Azure ecosystem integration: Data Factory, Power BI, Cosmos DB, Azure Storage, Azure ML, Purview, etc. Synapse aims to make learning and using the tools simple.

Databricks has a strong focus on Spark execution, Delta (Lakehouse Pattern) and ML lifecycles and enabling developers to have full control over their Spark environment.

Both have custom Spark runtimes. Databricks Runtime claims up to 50x performance gains over Apache Spark. Synapse Spark claims 2x - 18x improvement over Apache Spark. These are different implementations of the same APIs and it'd be worthwhile to understand the execution of your specific workloads on each for a true comparison.

Databricks has Spark 3.0. Synapse has Spark 2.4 (3.0 is currently in preview).

Both Databricks and Synapse support Notebook-based and local IDE based development (DBConnect + VS Code Synapse plugin).

Both support Data Lake integration using Delta.

In short, they are very competitive offerings.
