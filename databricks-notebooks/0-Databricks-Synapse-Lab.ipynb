# Lixar Databricks Synapse Lab

This lab will illustrate processes that span across Databricks and Synapse. It's primary goal is to show how Synapse can be incorporated into existing Databricks workflows.

## Agenda

### Setup

### Part 1: Background

* Examination of differences between Synapse and Databricks and their relative strengths. 
* Examples where you may prefer one over the other.

### Part 2: The Databricks Synapse Connector

* Writing data to Synapse tables
* Reading data from Synapse tables
* Streaming data to Synapse tables

### Part 3: Data Lake Interoperability

* Formats for interoperability
* Writing Delta in Databricks, Reading in Synapse
* Writing Delta in Synapse, Reading in Databricks
* Supporting SQL on-demand queries using parquet

### Part 4: Synapse Spark

* Illustrate how Spark can be used in both Databricks and Synapse on the same dataset

## Setup

The resources below should be prefixed / suffixed with a value to ensure the names are globally unique. For example, I'll use my initials snc to guarantee uniqueness.

The following resources are required prior to starting this lab:

1. Create a new Resource Group - `dbw-syn-workshop-<suffix>`
2. Create a new Synapse Workspace in the resource group - `syn-dbw-syn-workshop-<suffix>`
  * Create a new Storage Account for Synapse during creation - `strdbwsynworkshop<suffix>`
3. Create a new Databricks Workspace in the resource group - `dbw-dbw-syn-workshop-<suffix>`
4. A service principal w/ secret for Databricks (through App Registrations)- `dbw-syn-workshop-<suffix>`
5. A secret scope in databricks container the service principal secret, Synapse SQL password and Storage Account key

```
databricks secrets create-scope --scope dbw-syn-workshop --initial-manage-principal users
databricks secrets write --scope dbw-syn-lab --key dbw-syn-workshop-sp-secret --string-value <secret>
databricks secrets write --scope dbw-syn-lab --key dbw-syn-workshop-sa-secret --string-value <secret>
databricks secrets write --scope dbw-syn-lab --key dbw-syn-workshop-sql-pw --string-value <secret>
```
